```python
# -*- coding: utf-8 -*-
from py2neo import Node, Graph, Relationship, NodeMatcher
import pandas as pd
from tqdm import tqdm  # Import tqdm library for progress bar

class DataToNeo4j(object):
    """Load data from Excel into Neo4j"""

    def __init__(self, excel_path):
        """Establish connection"""
        # Neo4j connection
        link = Graph("http://localhost:7474", auth=("neo4j", "Yk1213903395!"))  # Ensure connection information is correct
        self.graph = link
        # Define node types (labels)
        self.Disease_1 = 'First'
        self.Disease_2 = 'Second'
        self.Disease_3 = 'Third'
        self.Disease_4 = 'Forth'
        self.Anatomical_location = 'Anatomical_location'
        self.Examination = 'Examination'
        self.OCT_sign = 'OCT_sign'
        self.Symptom = 'Symptom'
        self.Physical_sign = 'Physical_sign'
        self.Gene = 'Gene'
        self.Differential_diagnosis = 'Differential_diagnosis'
        self.Complication = 'Complication'
        self.Etiology = 'Etiology'
        self.Treatment_general = 'Treatment_general'
        self.Treatment_drug = 'Treatment_drug'
        self.Treatment_usage = 'Treatment_usage'
        self.Treatment_surgery = 'Treatment_surgery'
        self.Treatment_indications = 'Treatment_indications'
        self.Treatment_contraindications = 'Treatment_contraindications'
        self.Related_disease = 'Related_disease'
        self.Staging_typing = 'Staging_typing'
        self.Synonym_3 = 'Synonym_3'
        self.Synonym_4 = 'Synonym_4'
        self.Age_of_onset = 'Age_of_onset'
        self.High_risk_population = 'High_risk_population'
        self.Medical_history = 'Medical_history'

        self.graph.delete_all()  # Clear all content in the graph database
        self.matcher = NodeMatcher(link)

        # Read Excel data
        self.df_data = pd.read_excel(excel_path)
        self.df_data = self.df_data.applymap(
            lambda x: x.strip() if isinstance(x, str) else x
        )

        self.df_data.replace(
            ["", "entity", "entity class", "Entity", "Entity Class"],
            pd.NA,
            inplace=True
        )
        self.create_node(self.df_data['First'].tolist(),
                         self.df_data['Second'].tolist(),
                         self.df_data['Third'].tolist(),
                         self.df_data['Forth'].tolist(),
                         self.df_data['Anatomical_location'].tolist(),
                         self.df_data['Examination'].tolist(),
                         self.df_data['OCT_sign'].tolist(),
                         self.df_data['Symptom'].tolist(),
                         self.df_data['Physical_sign'].tolist(),
                         self.df_data['Gene'].tolist(),
                         self.df_data['Differential_diagnosis'].tolist(),
                         self.df_data['Complication'].tolist(),
                         self.df_data['Etiology'].tolist(),
                         self.df_data['Treatment_general'].tolist(),
                         self.df_data['Treatment_drug'].tolist(),
                         self.df_data['Treatment_usage'].tolist(),
                         self.df_data['Treatment_surgery'].tolist(),
                         self.df_data['Treatment_indications'].tolist(),
                         self.df_data['Treatment_contraindications'].tolist(),
                         self.df_data['Related_disease'].tolist(),
                         self.df_data['Staging_typing'].tolist(),
                         self.df_data['Synonym_3'].tolist(),
                         self.df_data['Synonym_4'].tolist(),
                         self.df_data['Age_of_onset'].tolist(),
                         self.df_data['High_risk_population'].tolist(),
                         self.df_data['Medical_history'].tolist(),)
        self.create_relation(self.df_data)

    def create_node(self, node_Disease_1_key, node_Disease_2_key, node_Disease_3_key, node_Disease_4_key, node_Anatomical_location_key, node_Examination_key, node_OCT_sign_key, node_Symptom_key, node_Physical_sign_key, node_Gene_key, node_Differential_diagnosis_key, node_Complication_key, node_Etiology_key, node_Treatment_general_key, node_Treatment_drug_key, node_Treatment_usage_key, node_Treatment_surgery_key, node_Treatment_indications_key, node_Treatment_contraindications_key, node_Related_disease_key, node_Staging_typing_key, node_Synonym_3_key, node_Synonym_4_key, node_Age_of_onset_key, node_High_risk_population_key, node_Medical_history_key):
        """Create nodes with MERGE to avoid duplicates and set custom IDs"""
        print("Creating nodes...")  # English message
        # Using tqdm to track progress
        for name in tqdm(node_Disease_1_key, desc="Creating First nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Disease_1_node = Node(self.Disease_1, name=name, id=abs(hash(name)) % 100000, label=self.Disease_1)  # Add label as property
            self.graph.merge(Disease_1_node, self.Disease_1, "name")

        for name in tqdm(node_Disease_2_key, desc="Creating Second nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Disease_2_node = Node(self.Disease_2, name=name, id=abs(hash(name)) % 100000, label=self.Disease_2)  # Add label as property
            self.graph.merge(Disease_2_node, self.Disease_2, "name")

        for name in tqdm(node_Disease_3_key, desc="Creating Third nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Disease_3_node = Node(self.Disease_3, name=name, id=abs(hash(name)) % 100000, label=self.Disease_3)  # Add label as property
            self.graph.merge(Disease_3_node, self.Disease_3, "name")

        for name in tqdm(node_Disease_4_key, desc="Creating Forth nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Disease_4_node = Node(self.Disease_4, name=name, id=abs(hash(name)) % 100000, label=self.Disease_4)  # Add label as property
            self.graph.merge(Disease_4_node, self.Disease_4, "name")

        for name in tqdm(node_Anatomical_location_key, desc="Creating Forth nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Anatomical_location_node = Node(self.Anatomical_location, name=name, id=abs(hash(name)) % 100000, label=self.Anatomical_location)  # Add label as property
            self.graph.merge(Anatomical_location_node, self.Anatomical_location, "name")

        for name in tqdm(node_Examination_key, desc="Creating Forth nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Examination_node = Node(self.Examination, name=name, id=abs(hash(name)) % 100000, label=self.Examination)  # Add label as property
            self.graph.merge(Examination_node, self.Examination, "name")

        for name in tqdm(node_OCT_sign_key, desc="Creating OCT_sign nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            OCT_sign_node = Node(self.OCT_sign, name=name, id=abs(hash(name)) % 100000, label=self.OCT_sign)  # Add label as property
            self.graph.merge(OCT_sign_node, self.OCT_sign, "name")

        for name in tqdm(node_Symptom_key, desc="Creating Symptom nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Symptom_node = Node(self.Symptom, name=name, id=abs(hash(name)) % 100000, label=self.Symptom)  # Add label as property
            self.graph.merge(Symptom_node, self.Symptom, "name")

        for name in tqdm(node_Physical_sign_key, desc="Creating Physical_sign nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Physical_sign_node = Node(self.Physical_sign, name=name, id=abs(hash(name)) % 100000, label=self.Physical_sign)  # Add label as property
            self.graph.merge(Physical_sign_node, self.Physical_sign, "name")

        for name in tqdm(node_Gene_key, desc="Creating Gene nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Gene_node = Node(self.Gene, name=name, id=abs(hash(name)) % 100000, label=self.Gene)  # Add label as property
            self.graph.merge(Gene_node, self.Gene, "name")

        for name in tqdm(node_Differential_diagnosis_key, desc="Creating Differential_diagnosis nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Differential_diagnosis_node = Node(self.Differential_diagnosis, name=name, id=abs(hash(name)) % 100000, label=self.Differential_diagnosis)  # Add label as property
            self.graph.merge(Differential_diagnosis_node, self.Differential_diagnosis, "name")

        for name in tqdm(node_Complication_key, desc="Creating Complication nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Complication_node = Node(self.Complication, name=name, id=abs(hash(name)) % 100000, label=self.Complication)  # Add label as property
            self.graph.merge(Complication_node, self.Complication, "name")

        for name in tqdm(node_Etiology_key, desc="Creating Etiology nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Etiology_node = Node(self.Etiology, name=name, id=abs(hash(name)) % 100000, label=self.Etiology)  # Add label as property
            self.graph.merge(Etiology_node, self.Etiology, "name")

        for name in tqdm(node_Treatment_general_key, desc="Creating Treatment_general nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Treatment_general_node = Node(self.Treatment_general, name=name, id=abs(hash(name)) % 100000, label=self.Treatment_general)  # Add label as property
            self.graph.merge(Treatment_general_node, self.Treatment_general, "name")
        
        for name in tqdm(node_Treatment_drug_key, desc="Creating Treatment_drug nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Treatment_drug_node = Node(self.Treatment_drug, name=name, id=abs(hash(name)) % 100000, label=self.Treatment_drug)  # Add label as property
            self.graph.merge(Treatment_drug_node, self.Treatment_drug, "name")
        
        for name in tqdm(node_Treatment_usage_key, desc="Creating Treatment_usage nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Treatment_usage_node = Node(self.Treatment_usage, name=name, id=abs(hash(name)) % 100000, label=self.Treatment_usage)  # Add label as property
            self.graph.merge(Treatment_usage_node, self.Treatment_usage, "name")

        for name in tqdm(node_Treatment_surgery_key, desc="Creating Treatment_surgery nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Treatment_surgery_node = Node(self.Treatment_surgery, name=name, id=abs(hash(name)) % 100000, label=self.Treatment_surgery)  # Add label as property
            self.graph.merge(Treatment_surgery_node, self.Treatment_surgery, "name")

        for name in tqdm(node_Treatment_indications_key, desc="Creating Treatment_indications nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Treatment_indications_node = Node(self.Treatment_indications, name=name, id=abs(hash(name)) % 100000, label=self.Treatment_indications)  # Add label as property
            self.graph.merge(Treatment_indications_node, self.Treatment_indications, "name")

        for name in tqdm(node_Treatment_contraindications_key, desc="Creating Treatment_contraindications nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Treatment_contraindications_node = Node(self.Treatment_contraindications, name=name, id=abs(hash(name)) % 100000, label=self.Treatment_contraindications)  # Add label as property
            self.graph.merge(Treatment_contraindications_node, self.Treatment_contraindications, "name")

        for name in tqdm(node_Related_disease_key, desc="Creating Related_disease nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Related_disease_node = Node(self.Related_disease, name=name, id=abs(hash(name)) % 100000, label=self.Related_disease)  # Add label as property
            self.graph.merge(Related_disease_node, self.Related_disease, "name")

        for name in tqdm(node_Staging_typing_key, desc="Creating Staging_typing nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Staging_typing_node = Node(self.Staging_typing, name=name, id=abs(hash(name)) % 100000, label=self.Staging_typing)  # Add label as property
            self.graph.merge(Staging_typing_node, self.Staging_typing, "name")

        for name in tqdm(node_Synonym_3_key, desc="Creating Synonym_3 nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Synonym_3_node = Node(self.Synonym_3, name=name, id=abs(hash(name)) % 100000, label=self.Synonym_3)  # Add label as property
            self.graph.merge(Synonym_3_node, self.Synonym_3, "name")

        for name in tqdm(node_Synonym_4_key, desc="Creating Synonym_4 nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Synonym_4_node = Node(self.Synonym_4, name=name, id=abs(hash(name)) % 100000, label=self.Synonym_4)  # Add label as property
            self.graph.merge(Synonym_4_node, self.Synonym_4, "name")

        for name in tqdm(node_Age_of_onset_key, desc="Creating Age_of_onset nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Age_of_onset_node = Node(self.Age_of_onset, name=name, id=abs(hash(name)) % 100000, label=self.Age_of_onset)  # Add label as property
            self.graph.merge(Age_of_onset_node, self.Age_of_onset, "name")

        for name in tqdm(node_High_risk_population_key, desc="Creating High_risk_population nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            High_risk_population_node = Node(self.High_risk_population, name=name, id=abs(hash(name)) % 100000, label=self.High_risk_population)  # Add label as property
            self.graph.merge(High_risk_population_node, self.High_risk_population, "name")

        for name in tqdm(node_Medical_history_key, desc="Creating Medical_history nodes", unit="node"):
            if pd.isna(name):  # If node name is empty, skip this node
                continue
            # Use a positive 5-digit integer ID
            Medical_history_node = Node(self.Medical_history, name=name, id=abs(hash(name)) % 100000, label=self.Medical_history)  # Add label as property
            self.graph.merge(Medical_history_node, self.Medical_history, "name")

    def create_relation(self, df_data):
        """Create relationships"""
        print("Creating relationships...")  # English message
        # Using tqdm to track progress
        for m in tqdm(range(0, len(df_data)), desc="Creating relationships", unit="relationship"):
            try:
                # Find nodes
                disease_1_name = df_data['First'][m]
                disease_2_name = df_data['Second'][m]
                disease_3_name = df_data['Third'][m]
                disease_4_name = df_data['Forth'][m]
                anatomical_location_name = df_data['Anatomical_location'][m]
                examination_name = df_data['Examination'][m]
                OCT_sign_name = df_data['OCT_sign'][m]
                symptom_name = df_data['Symptom'][m]
                physical_sign_name = df_data['Physical_sign'][m]
                gene_name = df_data['Gene'][m]
                differential_diagnosis_name = df_data['Differential_diagnosis'][m]
                complication_name = df_data['Complication'][m]
                etiology_name = df_data['Etiology'][m]
                treatment_general_name = df_data['Treatment_general'][m]
                treatment_drug_name = df_data['Treatment_drug'][m]
                treatment_usage_name = df_data['Treatment_usage'][m]
                treatment_surgery_name = df_data['Treatment_surgery'][m]
                treatment_indications_name = df_data['Treatment_indications'][m]
                treatment_contraindications_name = df_data['Treatment_contraindications'][m]
                related_disease_name = df_data['Related_disease'][m]
                staging_typing_name = df_data['Staging_typing'][m]
                synonym_3_name = df_data['Synonym_3'][m]
                synonym_4_name = df_data['Synonym_4'][m]
                age_of_onset_name = df_data['Age_of_onset'][m]
                high_risk_population_name = df_data['High_risk_population'][m]
                medical_history_name = df_data['Medical_history'][m]
                # Relationship
                contain_type = df_data['Contain'][m]
                anatomical_location_type = df_data['Located in'][m]
                examination_type = df_data['Requires examination'][m]
                oct_sign_type = df_data['Has oct sign'][m]
                gene_type = df_data['Related gene'][m]
                symptom_type = df_data['Has symptom'][m]
                physical_sign_type = df_data['Has physical sign'][m]
                differential_diagnosis_type = df_data['Needs distinguished from'][m]
                complication_type = df_data['May cause'][m]
                etiology_type = df_data['Caused by'][m]
                treatment_type = df_data['Treated with'][m]
                related_disease_type = df_data['Related to'][m]
                staging_typing_type = df_data['Classified as'][m]
                synonym_type = df_data['Same as'][m]
                age_of_onset_type = df_data['Onset during'][m]
                high_risk_population_type = df_data['Affects population'][m]
                medical_history_type = df_data['Related history'][m]

                # Skip row if any necessary data is missing
                if pd.isna(disease_1_name) or pd.isna(contain_type):
                    continue

                disease_1_node = self.matcher.match(self.Disease_1).where(f"_.name = '{disease_1_name}'").first()
                disease_2_node = self.matcher.match(self.Disease_2).where(f"_.name = '{disease_2_name}'").first() if not pd.isna(disease_2_name) else None
                disease_3_node = self.matcher.match(self.Disease_3).where(f"_.name = '{disease_3_name}'").first() if not pd.isna(disease_3_name) else None
                disease_4_node = self.matcher.match(self.Disease_4).where(f"_.name = '{disease_4_name}'").first() if not pd.isna(disease_4_name) else None
                OCT_sign_node = self.matcher.match(self.OCT_sign).where(f"_.name = '{OCT_sign_name}'").first() if not pd.isna(OCT_sign_name) else None
                Symptom_node = self.matcher.match(self.Symptom).where(f"_.name = '{symptom_name}'").first() if not pd.isna(symptom_name) else None
                Anatomical_location_node = self.matcher.match(self.Anatomical_location).where(f"_.name = '{anatomical_location_name}'").first() if not pd.isna(anatomical_location_name) else None
                Examination_node = self.matcher.match(self.Examination).where(f"_.name = '{examination_name}'").first() if not pd.isna(examination_name) else None
                Physical_sign_node = self.matcher.match(self.Physical_sign).where(f"_.name = '{physical_sign_name}'").first() if not pd.isna(physical_sign_name) else None
                Gene_node = self.matcher.match(self.Gene).where(f"_.name = '{gene_name}'").first() if not pd.isna(gene_name) else None
                Differential_diagnosis_node = self.matcher.match(self.Differential_diagnosis).where(f"_.name = '{differential_diagnosis_name}'").first() if not pd.isna(differential_diagnosis_name) else None
                Complication_node = self.matcher.match(self.Complication).where(f"_.name = '{complication_name}'").first() if not pd.isna(complication_name) else None
                Etiology_node = self.matcher.match(self.Etiology).where(f"_.name = '{etiology_name}'").first() if not pd.isna(etiology_name) else None
                Treatment_general_node = self.matcher.match(self.Treatment_general).where(f"_.name = '{treatment_general_name}'").first() if not pd.isna(treatment_general_name) else None
                Treatment_drug_node = self.matcher.match(self.Treatment_drug).where(f"_.name = '{treatment_drug_name}'").first() if not pd.isna(treatment_drug_name) else None
                Treatment_usage_node = self.matcher.match(self.Treatment_usage).where(f"_.name = '{treatment_usage_name}'").first() if not pd.isna(treatment_usage_name) else None
                Treatment_surgery_node = self.matcher.match(self.Treatment_surgery).where(f"_.name = '{treatment_surgery_name}'").first() if not pd.isna(treatment_surgery_name) else None
                Treatment_indications_node = self.matcher.match(self.Treatment_indications).where(f"_.name = '{treatment_indications_name}'").first() if not pd.isna(treatment_indications_name) else None
                Treatment_contraindications_node = self.matcher.match(self.Treatment_contraindications).where(f"_.name = '{treatment_contraindications_name}'").first() if not pd.isna(treatment_contraindications_name) else None
                Related_disease_node = self.matcher.match(self.Related_disease).where(f"_.name = '{related_disease_name}'").first() if not pd.isna(related_disease_name) else None
                Staging_typing_node = self.matcher.match(self.Staging_typing).where(f"_.name = '{staging_typing_name}'").first() if not pd.isna(staging_typing_name) else None
                Synonym_3_node = self.matcher.match(self.Synonym_3).where(f"_.name = '{synonym_3_name}'").first() if not pd.isna(synonym_3_name) else None
                Synonym_4_node = self.matcher.match(self.Synonym_4).where(f"_.name = '{synonym_4_name}'").first() if not pd.isna(synonym_4_name) else None
                Age_of_onset_node = self.matcher.match(self.Age_of_onset).where(f"_.name = '{age_of_onset_name}'").first() if not pd.isna(age_of_onset_name) else None
                High_risk_population_node = self.matcher.match(self.High_risk_population).where(f"_.name = '{high_risk_population_name}'").first() if not pd.isna(high_risk_population_name) else None
                Medical_history_node = self.matcher.match(self.Medical_history).where(f"_.name = '{medical_history_name}'").first() if not pd.isna(medical_history_name) else None

                # Step 1: Connect disease hierarchy
                if disease_2_node:
                    # Connect disease 1 and disease 2
                    if disease_1_node and disease_2_node:
                        rel = Relationship(disease_1_node, contain_type, disease_2_node)
                        self.graph.create(rel)
                    # Connect disease 2 and disease 3
                    if disease_2_node and disease_3_node:
                        rel = Relationship(disease_2_node, contain_type, disease_3_node)
                        self.graph.create(rel)
                else:
                    # Connect disease 1 and disease 3 directly if disease 2 doesn't exist
                    if disease_1_node and disease_3_node:
                        rel = Relationship(disease_1_node, contain_type, disease_3_node)
                        self.graph.create(rel)

                # Step 2: Handle disease 4 and synonym relationships
                if disease_4_node:
                    # Connect disease 3 and disease 4
                    if disease_3_node and disease_4_node:
                        rel = Relationship(disease_3_node, contain_type, disease_4_node)
                        self.graph.create(rel)
                    
                    # Connect disease 3 and synonym 3
                    if disease_3_node and Synonym_3_node:
                        rel = Relationship(disease_3_node, synonym_type, Synonym_3_node)
                        self.graph.create(rel)
                    
                    # Connect disease 4 and synonym 4
                    if disease_4_node and Synonym_4_node:
                        rel = Relationship(disease_4_node, synonym_type, Synonym_4_node)
                        self.graph.create(rel)
                    
                    # Set target node to disease 4 for other relationships
                    target_1_node = disease_4_node
                else:
                    # Connect disease 3 and synonym 3 only
                    if disease_3_node and Synonym_3_node:
                        rel = Relationship(disease_3_node, synonym_type, Synonym_3_node)
                        self.graph.create(rel)
                    
                    # Set target node to disease 3 for other relationships
                    target_1_node = disease_3_node

                # Step 3: Handle staging and typing relationships
                if Staging_typing_node:
                    # Connect target_1_node and anatomical
                    if target_1_node and Staging_typing_node:
                        rel = Relationship(target_1_node, staging_typing_type, Staging_typing_node)
                        self.graph.create(rel)
                    target_2_node = Staging_typing_node
                else:
                    # Connect disease and other content directly if anatomical exist
                    target_2_node = target_1_node

                # Create relationships for other entities based on the target node
                if target_2_node:
                    if Anatomical_location_node:
                        rel = Relationship(target_2_node, anatomical_location_type, Anatomical_location_node)
                        self.graph.create(rel)
                    if Examination_node:
                        rel = Relationship(target_2_node, examination_type, Examination_node)
                        self.graph.create(rel)
                    if OCT_sign_node:
                        rel = Relationship(target_2_node, oct_sign_type, OCT_sign_node)
                        self.graph.create(rel)
                    if Gene_node:
                        rel = Relationship(target_2_node, gene_type, Gene_node)
                        self.graph.create(rel)
                    if Related_disease_node:
                        rel = Relationship(target_2_node, related_disease_type, Related_disease_node)
                        self.graph.create(rel)
                    if Treatment_general_node:
                        rel = Relationship(target_2_node, treatment_type, Treatment_general_node)
                        self.graph.create(rel)
                    if Treatment_drug_node:
                        rel = Relationship(target_2_node, treatment_type, Treatment_drug_node)
                        self.graph.create(rel)
                    if Treatment_usage_node:
                        rel = Relationship(target_2_node, treatment_type, Treatment_usage_node)
                        self.graph.create(rel)
                    if Treatment_surgery_node:
                        rel = Relationship(target_2_node, treatment_type, Treatment_surgery_node)
                        self.graph.create(rel)
                    if Treatment_indications_node:
                        rel = Relationship(target_2_node, treatment_type, Treatment_indications_node)
                        self.graph.create(rel)
                    if Treatment_contraindications_node:
                        rel = Relationship(target_2_node, treatment_type, Treatment_contraindications_node)
                        self.graph.create(rel)
                    if Etiology_node:
                        rel = Relationship(target_2_node, etiology_type, Etiology_node)
                        self.graph.create(rel)
                    if Complication_node:
                        rel = Relationship(target_2_node, complication_type, Complication_node)
                        self.graph.create(rel)
                    if Differential_diagnosis_node:
                        rel = Relationship(target_2_node, differential_diagnosis_type, Differential_diagnosis_node)
                        self.graph.create(rel)
                    if Physical_sign_node:
                        rel = Relationship(target_2_node, physical_sign_type, Physical_sign_node)
                        self.graph.create(rel)
                    if Symptom_node:
                        rel = Relationship(target_2_node, symptom_type, Symptom_node)
                        self.graph.create(rel)
                    if Age_of_onset_node:
                        rel = Relationship(target_2_node, age_of_onset_type, Age_of_onset_node)
                        self.graph.create(rel)
                    if High_risk_population_node:
                        rel = Relationship(target_2_node, high_risk_population_type, High_risk_population_node)
                        self.graph.create(rel)
                    if Medical_history_node:
                        rel = Relationship(target_2_node, medical_history_type, Medical_history_node)
                        self.graph.create(rel)

            except AttributeError as e:
                print(f"Error at index {m}: {e}")

if __name__ == "__main__":
    # Pass Excel file path as parameter
    excel_path = "D:\\Data\\Graduate_Content\\Ophthalmology_Papers\\Artificial_intelligence\\Knowledge_Graph\\Retinal_Disease_Knowledge_Graph_Building\\04 Neo4j_Building\\Database.xlsx"
    excel_data = DataToNeo4j(excel_path)
```
